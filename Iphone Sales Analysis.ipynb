```python
# iPhone Sales Data Analysis

In this notebook, we will analyze the iPhone sales dataset using Python. The dataset contains information about various iPhone models, including their sale prices, ratings, and RAM specifications.

### Objectives
1. **Data Loading and Inspection**: Load the dataset and examine its structure, summary statistics, and any missing values.
2. **Data Cleaning**: Perform necessary data cleaning steps to ensure data quality for analysis.
3. **Analysis and Visualization**:
   - **Model with the Highest Rating**: Identify the iPhone model with the highest rating.
   - **Sale Price Distribution**: Visualize the distribution of sale prices across different iPhone models.
   - **Ratings and Number of Ratings**: Examine the correlation between the number of ratings and star ratings, and visualize this relationship.
   - **Average RAM Size**: Calculate and display the average RAM size of the iPhones.
   - **Price vs. Star Rating**: Analyze and visualize the relationship between the sale price and star rating of the iPhones.

The analysis will be performed using the pandas, numpy, and plotly libraries.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load the dataset
file_path = "C:/Users/lenovo/Desktop/Husain/Data Science Projects/IPhone Sales Analysis/apple_products.csv"
data = pd.read_csv(file_path)

# Display the first few rows of the data
print(data.head())

# Check the summary statistics
print(data.describe())

# Check for missing values
print(data.isnull().sum())

# Check data types
print(data.dtypes)

# Data Cleaning: Dropping missing values in important columns
data.dropna(subset=['Sale Price', 'Star Rating', 'Number Of Ratings'], inplace=True)

# Convert 'Ram' column to integer after removing 'GB'
data['Ram'] = data['Ram'].str.replace('GB', '').astype(int)

# Analysis and Visualization

# A. Find the iPhone model with the highest rating
highest_rating_model = data.loc[data['Star Rating'].idxmax()]
print("Model with the highest rating:")
print(highest_rating_model)

# B. Distribution of Sale Prices Across iPhone Models
fig1 = px.bar(
    data,
    x='Product Name',
    y='Sale Price',
    title='Sale Price Distribution Across iPhone Models',
    labels={'Product Name': 'iPhone Model', 'Sale Price': 'Sale Price'},
    height=600,
    width=1000
)
fig1.update_xaxes(title_text='iPhone Model', tickangle=45)
fig1.update_yaxes(title_text='Sale Price')
fig1.show()

# C. Correlation Between Number of Ratings and Star Rating
correlation = data['Number Of Ratings'].corr(data['Star Rating'])
print(f"Correlation between Number of Ratings and Star Rating: {correlation}")

# Scatter plot for Number of Ratings vs Star Rating
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Number Of Ratings', y='Star Rating', data=data)
plt.title('Number of Ratings vs Star Rating')
plt.xlabel('Number Of Ratings')
plt.ylabel('Star Rating')
plt.show()

# D. Average RAM Size of iPhones
average_ram = data['Ram'].mean()
print(f"Average RAM size of the iPhones: {average_ram} GB")

# E. Price vs. Star Rating Analysis
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Sale Price', y='Star Rating', data=data)
plt.title('Price vs. Star Rating')
plt.xlabel('Sale Price')
plt.ylabel('Star Rating')
plt.show()
```

### Insights from iPhone Sales Data Analysis

1. **Highest Rated Model**:
   - The iPhone model with the highest rating is the **APPLE iPhone 11 Pro Max (Gold, 256 GB)** with a star rating of 4.7. This model has the highest sale price of ₹131,900 and does not offer any discount.

2. **Sale Price Distribution**:
   - The distribution of sale prices across different iPhone models shows a significant variation. The median sale price is ₹75,900, but prices can range from ₹29,999 to ₹1,40,900, indicating a wide range of models with varying features and price points.

3. **Correlation Between Ratings and Number of Ratings**:
   - There is a weak negative correlation (-0.225) between the number of ratings and the star rating. This suggests that higher-rated products do not necessarily have a higher number of ratings. The correlation is not strong, so other factors may be influencing ratings.

4. **Average RAM Size**:
   - The average RAM size of the iPhones in this dataset is approximately **4.18 GB**. This indicates that most iPhones have a relatively high RAM capacity, which is expected in high-end models.

5. **Price vs. Star Rating**:
   - The scatter plot of price vs. star rating shows a general trend where higher prices do not always correspond to higher ratings. This suggests that while price may reflect the model’s features, it does not necessarily guarantee a better rating.
